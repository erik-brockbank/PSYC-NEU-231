```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from pandas import DataFrame

tips = sns.load_dataset('tips')
tips = pd.DataFrame(data = tips)
print(tips.shape)
# tips.head()

# 1.
tips["tip_pct"] = tips["tip"] / tips["total_bill"]
tips.head()

# 2.
plot_days = sns.pointplot(x = "day", y = "tip_pct", hue = "time", data = tips)

# 3.
plot_party = sns.catplot(x = "day", y = "tip_pct", hue = "time", col = "size", data = tips, kind = "point")

plt.show()
```
